# Підгрупи, ізоморфні групи, системи твірних

## 1. Системи твірних

```python
G = DihedralGroup(10)

g1 = G.list()[1]
g2 = G.list()[2]
print(g1, g2, sep='\n')
```

```python
H1 = G.subgroup([g1, g2])
H1
```

```python
H1.is_cyclic()
```

```python
H1.order()
```

```python
# таким чином можна перевірити чи підгрупа рівна всій групі 
H1 == G
```

```python
g1 = G.list()[5]
g2 = G.list()[6]
g3 = G.list()[10]
print(g1, g2, g3, sep='\n')
H = G.subgroup([g1, g2, g3])
print(H)
print('subgroup isomorphic to G:', H == G)
```

```python
H.cayley_graph().plot(layout='circular', vertex_labels=False,)
```

```python
G.cayley_graph().plot()
```

## 2. Ізоморфні групи 

Для груп, заданих підстановками, можна використати метод `structure_description` і, відповідно, метод `is_isomorphic`. Хоча не всі скінченні групи в сейджі мають представлення як підгрупи груп підстановок. 

```python
G1 = CyclicPermutationGroup(7)
G1.structure_description()
```

```python
G2 = SymmetricGroup(5)
G2.structure_description()
```

```python
g1 = G2.list()[4]
g2 = G2.list()[7]

H = G2.subgroup([g1, g2])
H
```

```python
H.structure_description()
```

```python
G3 = DihedralGroup(12)
H = G3.subgroups()[20]
H
```

```python
H.structure_description()
```

```python
H.is_isomorphic(CyclicPermutationGroup(4))
```

```python
G4 = SL(2, Integers(3))
print(G4)
print('order:', G4.order())
print('as permutation group:', G4.as_permutation_group())
print('structure description:', G4.structure_description())
```

```python
# метод працює тільки для підстановочних груп
G4.is_isomorhic(DihedralGroup(12))
```
